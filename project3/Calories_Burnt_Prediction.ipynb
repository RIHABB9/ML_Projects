```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing and splitting
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Models
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

# Evaluation
from sklearn import metrics

import warnings
warnings.filterwarnings('ignore')
```

```python
# Step 2: Load datasets (+ merge them) into pd dataframe & explore it
```

```python
df_calories = pd.read_csv('calories.csv')
df_exercise = pd.read_csv('exercise.csv')

# merge the 2 dataframes
df = pd.merge(df_exercise, df_calories, how = 'outer')
df.head()
```

```python
df.replace({'male': 0, 'female': 1}, inplace=True)
df.head()
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(df.corr() > 0.9, annot=True, cbar=False)
plt.show()
```

```python
to_remove = ['Weight', 'Duration']
df.drop(to_remove, axis=1, inplace=True)
```

```python
# Step 4: Model training
```

```python
features = df.drop(['User_ID', 'Calories'], axis=1)
target = df['Calories'].values

X_train, X_val, Y_train, Y_val = train_test_split(features, target, test_size=0.1, random_state=22)
X_train.shape, X_val.shape
```

```python
# normalizing the data for stable and fast training
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
from sklearn.metrics import mean_absolute_error as mae
models = [LinearRegression(), XGBRegressor(), Lasso(), RandomForestRegressor(), Ridge()]

for i in range(5):
    models[i].fit(X_train, Y_train)

    print(f'{models[i]} : ')

    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(Y_train, train_preds))

    val_preds = models[i].predict(X_val)
    print('Validation Error : ', mae(Y_val, val_preds))
    print()
```
